```python
import tensorflow as tf

print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))
```

```python
# upload cleaned file
import pandas as pd
import numpy as np

# import the data
data = pd.read_excel('cleaned_final.xlsx')
data
```

```python
def remove_xx(strings):
    return [s.replace("xx", "") for s in strings]

modified_strings = remove_xx(data['zip_code'])


data['zip_code'] = modified_strings
data['zip_code'] = data['zip_code'].astype(int)

type(data['zip_code'])
```

```python
type(data['loan_is_bad'])
```

```python
features_data = data.drop(columns=['loan_is_bad']).values
target_variable = data['loan_is_bad'].values

tensors = tf.data.Dataset.from_tensor_slices((features_data, target_variable))
```

```python
data_size = len(data)
print(data_size)
```

```python
train_dataset = tensors.take(round(len(data)*0.8),0)
test_dataset = tensors.skip(round(len(data)*0.8),0)
len(train_dataset)
```

```python
train_batch = train_dataset.batch(400) # batch size = 400
features, labels = next(iter(train_batch)) # iterate through each batch at training time

# print the first 20 records (rows) of features and labels
print(features[0:20])
print(labels[0:20])
```

```python
model = tf.keras.Sequential([
  tf.keras.layers.Dense(5, activation=tf.nn.relu, input_shape=(17, )),  # 10 features
  tf.keras.layers.Dense(5, activation=tf.nn.relu),
  tf.keras.layers.Dense(1)
])
```

```python
# compile the model
model.compile(optimizer=tf.keras.optimizers.AdamW(learning_rate=0.001,  weight_decay=0.005),
              loss=tf.keras.losses.MeanSquaredError(),
              metrics=['mse'])

model.fit(features, labels, epochs=2000)
```
